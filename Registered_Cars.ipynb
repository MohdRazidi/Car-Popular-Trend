```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Registered_Cars_Merged/cars_2023_2024_2025_merged.csv")

df.head()
```

```python
# Count values in each categorical column
print(df['fuel'].value_counts())
```

```python
# Update any fuel containing 'diesel' to 'Diesel'
df['fuel'] = df['fuel'].replace(
    to_replace = r'.*diesel.*',  # regex pattern
    value = 'Diesel',
    regex=True
)
```

```python
df['fuel'] = df['fuel'].replace({'petrol_ng': 'petrol'})
```

```python
df['fuel'] = df['fuel'].str.capitalize().str.strip()
```

```python
# Count values in each categorical column
print(df['fuel'].value_counts())
```

```python
df[df['fuel'].str.lower() == 'other']
```

```python
# Create a lookup table: for each maker-model-type, what is the most common fuel?
fuel_map = (
    df[df['fuel'] != 'Other']
    .groupby(['maker', 'model', 'type'])['fuel']
    .agg(lambda x: x.value_counts().index[0])
)

# Function to replace fuel='other' using the groups common fuel
def fix_other(row):
    if row['fuel'] == 'Other':
        key = (row['maker'], row['model'], row['type'])
        return fuel_map.get(key, 'Other')  # fallback 'other' if no match
    return row['fuel']

df['fuel'] = df.apply(fix_other, axis=1)
```
